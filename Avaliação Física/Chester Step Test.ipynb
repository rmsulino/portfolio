```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

fig, ax = plt.subplots(figsize=(12,6))
ax.axis([10, 40, 60, 200])


# X (varia de acordo com o tamanho do step)

X = [12, 17, 21, 25]

# Y = frequencias cardiacas obtidas em cada n√≠vel do teste

Y = [102, 111, 133, 151]


# Fazer para 3, 4 e 5 medidas

a = 4*((Y[0]*X[0])+(Y[1]*X[1])+(Y[2]*X[2])+(Y[3]*X[3]))
print(a)
b = (Y[0]+Y[1]+Y[2]+Y[3])*(X[0]+X[1]+X[2]+X[3])
print(b)
c = 4*((Y[0]*Y[0])+(Y[1]*Y[1])+(Y[2]*Y[2])+(Y[3]*Y[3]))
print(c)
d = (Y[0]+Y[1]+Y[2]+Y[3])*(Y[0]+Y[1]+Y[2]+Y[3])
print(d)
m = round((a-b)/(c-d),2)
print(m)
e = X[0]+X[1]+X[2]+X[3]
print(e)
f = round(m*(Y[0]+Y[1]+Y[2]+Y[3]))
print(f)
g = round((e-f)/4, 2)
print(g)
fcmax = 179

VO2max = round(m*fcmax+g, 2)
print('VO2max =', VO2max, 'ml/kg/min')

X.append(VO2max)
Y.append(fcmax)


p1 = np.polyfit(X, Y, 1)

maxHR = 179
HR80 = round(maxHR*0.8)
print('80% FC max = ', HR80)

# solve for a and b
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

    return a, b

# solution
a, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line

plt.scatter(X, Y, color='black')
xlims = plt.xlim([10, 38])
X.insert(0, xlims[0])
Y.insert(0, np.polyval(p1, xlims[0]))
X.append(xlims[1])
Y.append(np.polyval(p1, xlims[1]))

yfit = [a + b * xi for xi in X]
plt.plot(X, yfit)
plt.xlim(xlims)
plt.axhline(y=maxHR, color='r', linestyle='-', linewidth = 1.5)
plt.axhline(y=HR80, color='r', linestyle='-', linewidth = 1.5)



plt.axvline(x=VO2max,color='green', linestyle='-', linewidth = 2)

plt.xticks(np.arange(10, 40, 2))

resultado = str('VO${_2}$max = '+str(VO2max)+' ml/kg/min')

texto_fcmax = 'FC max = '+str(maxHR)+' bpm'
texto_fc80 = '80% FC max = '+str(HR80)+' bpm'

ax.text(21, 85, resultado, transform=ax.transData, fontsize='16', weight='normal', fontstyle='italic')
ax.text(max(X)+.5, maxHR, texto_fcmax, transform=ax.transData, fontsize='10', color='black')
ax.text(max(X)+.5, HR80, texto_fc80, transform=ax.transData, fontsize='10', color='black')
ax.grid()

plt.show()
```
