# Somatotipo

```r
library(repr)

x <- c(0,-6,6,0,-4.5,4.5,0)
y <- c(12,-6,-6,0,4.5,4.5,-7.5)
par(mar = c(2,0,0,2), mgp = c(2,1,0))
plot(x,y,pch = 20,xlab = " ",ylab = " ",xlim = c(-8,8), ylim = c(-10,16),las = 1, col = "white",axes = F)
axis(4, las = 1, yaxp = c(-10,16,13),cex.axis=0.8)
axis(1, xaxp = c(-8, 8, 16),cex.axis=0.8)
# Segmentes
segments(x0 = 0, y0=-7.5, x1 = 0, y1 = 12, lty = 2)
segments(x0 = -6, y0=-6, x1 = 4.5, y1 = 4.5, lty = 2)
segments(x0 = 6, y0=-6, x1 = -4.5, y1 = 4.5, lty = 2)
# text
#windowsFonts(B=windowsFont("Bookman Old Style"))
text(0,13,"MESOMORFO", cex = 1.6,family="B", font = 2)
text(-6,-8,"ENDOMORFO",cex = 1.6,family="B", font = 2)
text(6,-8,"ECTOMORFO", cex = 1.6,family="B", font = 2)

xspline(y = c(-6, 4.5, 12), x =  c(-6, -4.5, 0), shape = -1, lty = 2)
xspline(y = c(-6, -7.5,-6), x =  c(-6, 0, 6), shape = -1, lty = 2)
xspline(y = c(-6, 4.5, 12), x = c(6, 4.5, 0),  shape = -1, lty = 2)


# Realizar o calculo (dados de exemplo)

estatura = 172 # cm
massa_corporal = 108 # kg
dobra_triceps = 3.5 # cm
dobra_subscapular = 6.5 # cm
dobra_suprailiaca = 4.0 # cm
dobra_panturrilha_medial = 6.5 # cm
biepicondilar_umero = 6.4 # cm
biepicondilar_femur = 8.0 # cm
circunf_biceps = 30 # cm 
circunf_panturrilha = 35 # cm

X = (dobra_triceps + dobra_subscapular + dobra_suprailiaca) * (170.18 / estatura)

endomorfo = - 0.7182 + 0.1451 * (X) - 0.00068 * (X^2) + 0.0000014 * (X^3)

mesomorfo = 0.858 * biepicondilar_umero + 0.601 * biepicondilar_femur + 0.188 * circunf_biceps + 0.161 * circunf_panturrilha - estatura * 0.131 + 4.5

hwr = estatura / (massa_corporal^0.3333)

if (hwr >= 40.75) {
    ectomorfo = 0.732 * hwr - 28.58
    }

if (hwr < 40.75 & hwr >= 38.25) {
    ectomorfo = 0.463 * hwr - 17.63
    }
    
if (hwr < 38.25) {
    ectomorfo = 0.1
    }

ponto_x = ectomorfo - endomorfo

ponto_y = 2 * mesomorfo - (endomorfo + ectomorfo)

points(ponto_x,ponto_y) #resultado do calculo

#text(ponto_x+0.3, ponto_y, labels='xxx')


# Visualizar

options(device="png")
options(repr.plot.width=15, repr.plot.height=15)

png('somatotipo.png', width=800, height=600, res=300)
```
